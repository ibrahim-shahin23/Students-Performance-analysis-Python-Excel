```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
df.head()
```

```python
df.shape
```

```python
# Count the number of null values in each column
null_counts = df.isnull().sum()

print("Null values in each column:")
print(null_counts)
```

```python
df = df.dropna(how="any")
df.shape
```

```python
for column_name in ["Parental_Involvement",  "Access_to_Resources", "Motivation_Level" ,  "Family_Income" , "Teacher_Quality"]:
    print(df[column_name].value_counts())
```

```python
for column_name in ["Parental_Involvement",  "Access_to_Resources", "Motivation_Level" ,  "Family_Income" , "Teacher_Quality"]:
    df[column_name] = df[column_name].replace({"Low":0,"Medium":1,"High":2})
df.head()
```

```python
for column_name in ["Extracurricular_Activities","Internet_Access","Learning_Disabilities"]:
    print(df[column_name].value_counts()) 
```

```python
for column_name in ["Extracurricular_Activities","Internet_Access","Learning_Disabilities"]:
    df[column_name] = df[column_name].replace({"No":0,"Yes":1})
df.head()
```

```python
df['School_Type'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
def one_hot_encode(data_frame , column_name):
    encoder = OneHotEncoder(sparse_output=False)
    column_to_encode = data_frame[[column_name]]    
    encoded_columns = encoder.fit_transform(column_to_encode)

    feature_names = encoder.get_feature_names_out()
    encoded_columns = pd.DataFrame(encoded_columns, columns=feature_names)

    df = data_frame.drop(column_name, axis=1)
    return encoded_columns
```

```python

encoded_columns = one_hot_encode(df , 'School_Type')
df = pd.concat([df.iloc[:, :12], encoded_columns, df.iloc[:, 13:]], axis=1)
df
```

```python
encoded_columns = one_hot_encode(df , 'Gender')
df = pd.concat([df.iloc[:, :19], encoded_columns, df.iloc[:,20]], axis=1)
df
```

```python
df['Peer_Influence'].value_counts()
```

```python
df['Parental_Education_Level'].value_counts()
```

```python
df['Distance_from_Home'].value_counts()
```

```python
df['Distance_from_Home'] = df['Distance_from_Home'].replace({"Near":0,"Moderate":1,"Far":2})
df['Parental_Education_Level'] = df['Parental_Education_Level'].replace({"High School":0,"College":1,"Postgraduate":2})
df['Peer_Influence'] = df['Peer_Influence'].replace({"Negative":-1,"Neutral":0,"Positive":1})
df
```

```python
import seaborn as sns
plt.figure(figsize=(14, 10))
```

```python
columns_to_exclude = ['School_Type_Private', 'School_Type_Public','Gender_Female',
       'Gender_Male',]

# Print the DataFrame excluding the specified columns
plt.figure(figsize=(14, 10))
sns.heatmap(round(df.drop(columns=columns_to_exclude).corr(),2),cbar=True,annot=True)
```

```python
corr_matrix = df.drop(columns=columns_to_exclude).corr()

# Filter the correlation matrix to keep only values > 0.1
filtered_corr_matrix = corr_matrix[(1 > corr_matrix.abs()) & (corr_matrix.abs()>0.1)]

# Plot the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(round(filtered_corr_matrix, 2), cbar=True, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```
